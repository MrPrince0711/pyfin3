```python
from pandas_datareader import data as pdr
import datetime as dt
import pandas as pd
import yfinance as yf 
import numpy as np
```

```python
yf.pdr_override()
```

```python
end = dt.datetime.now()
start = dt.datetime(2020,1,1)
```

```python
stockList = ['FSR', 'SBK','NED','ABG' ]
stocks = ['^J580.JO'] + [i + '.JO' for i in stockList]
stocks
```

```python
df = pdr.get_data_yahoo(stocks, start, end)
df
```

```python
log_returns = np.log(df.Close / df.Close.shift(1)).dropna()
log_returns.head()
```

```python
def calc_beta(df):
    np_array = df.values
    m = np_array[:,4]
    beta = []
    for ind, col in enumerate(df): 
        if ind < 4:
            s = np_array[:,ind]
            covariance = np.cov(s,m)
            beta.append( covariance[0,1]/covariance[1,1] )
    
    return pd.Series(beta, df.columns[:4], name = 'Beta')
```

```python
beta = calc_beta(log_returns)
beta
```

```python
units = np. array([200, 50, 350, 250,])

JSE_fin_prices = df.Close[-1:].values.tolist()[0]
price_cents = np.array([round(price,2) for price in JSE_fin_prices[1:]])
price_rands = price_cents/100
value = units*price_rands
weight = [round(val/sum(value),2) for val in value]
beta = round(beta, 2)
```

```python
port = pd.DataFrame({
    'Stock': stockList,
    'Type': 'Shares',
    'Price':  price_rands,
    'NO. shares': units,
    'Value': value,
    'Weights': weight,
    'Beta': beta,
    'Weighted Beta': weight*beta,
})
```
